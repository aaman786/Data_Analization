```python
from matplotlib import pyplot as plt
plt.plot([1,2,3],[2,4,5])
plt.show()
```

```python
#plotting the single data.
plt.plot([1,2,3],[2,4,5])
plt.title("Aman's data") 
plt.ylabel("this is y")
plt.xlabel("this is x")
plt.show()
```

```python
# for plotting two Datas
from matplotlib import pyplot as plt
x= [1,2,3]
y=[2,4,6]
x2=[1,4,2]
y2=[3,6,7]
plt.plot(x,y, label="first", linewidth=8)
plt.plot(x2,y2, label ="second")
plt.legend() #function for showing the labels
plt.show()
```

```python
# For changing plotting style
from matplotlib import style
style.use('ggplot') #using plotting style
x= [1,2,3]
y=[2,4,6]
x2=[1,4,2]
y2=[3,6,7]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```
